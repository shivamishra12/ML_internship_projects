```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as  sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import Ridge,Lasso
from sklearn.metrics import r2_score,mean_squared_error
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('housePredictionDataSet.csv',delim_whitespace=True)
data
```

```python
data.info()
```

```python
data.columns = ["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT","MEDV"]
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
x = data.iloc[:,:-1]
y= data.iloc[:,-1]
x
```

```python
y
```

```python
x_scaler = StandardScaler()
x_scaler = x_scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaler,y,test_size = 0.2,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
ridge = Ridge()
```

```python
param_grid = {
    "alpha": [0.001, 0.01, 0.1, 1, 10, 50, 100, 200, 500]
}

model = GridSearchCV(estimator=ridge,param_grid=param_grid,cv=5,scoring='r2',n_jobs=-1)
```

```python
model = model.best_estimator_
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
r2_score(y_pred,y_test)
```

```python
mean_squared_error(y_pred,y_test)
```
